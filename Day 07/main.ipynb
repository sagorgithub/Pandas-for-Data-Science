# Conditional Statements

```python
df.head()  # Display the first few rows of the DataFrame
```

```python
df['cnt'].max()  # Maximum value in the 'cnt' column
```

```python
df['cnt'] > 15
```

```python
df[df['cnt'] > 15] # Filter rows where 'cnt' is greater than 15
```

```python
# subseting rowa
df[df['dteday'] == '1/1/2011']  # Filter rows where 'dteday' is '1/1/2011'
```

```python
# subseting rowa
df[df['dteday'] == '12/31/2012']  # Filter rows where 'dteday' is '12/31/2012'
```

```python
df[(df['dteday'] == '1/1/2011') & (df['cnt'] > 15)]  # Filter rows where 'dteday' is '1/1/2011' and 'cnt' is greater than 15
```

```python
df[(df['dteday'] == '1/1/2011') | (df['dteday'] == '12/31/2012')]  # Filter rows where 'dteday' is either '1/1/2011' or '12/31/2012'
```

```python
df['dteday'].isin(['1/1/2011', '12/31/2012'])  # Check if 'dteday' is in the specified list
```

```python
df[df['dteday'].isin(['1/1/2011', '12/31/2012'])]  # Check if 'dteday' is in the specified list
```
